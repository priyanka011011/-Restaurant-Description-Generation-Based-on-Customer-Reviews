```python
import numpy as np
import pandas as pd
import re
import pandas as pd
import matplotlib.pyplot as plt
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv('/content/Restaurant_Corrected_Review.csv')
```

```python
df.head()
```

```python
# Convert the 'review' column to lowercase
df['cor_lem_review'] = df['cor_lem_review'].str.lower()
```

```python
# Define a function to filter out short sentences
def filter_short_sentences(text):
    # Split the text into sentences
    sentences = text.split('. ')  # Split by sentence delimiter

    # Filter out sentences with 3 words or fewer
    filtered_sentences = [sentence for sentence in sentences if len(sentence.split()) > 3]

    # Join the filtered sentences back into a text
    filtered_text = '. '.join(filtered_sentences)

    return filtered_text
```

```python
# Apply the filtering function to the 'reviews' column
df['cor_lem_review'] = df['cor_lem_review'].apply(filter_short_sentences)
```

```python
# Initialize the sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to calculate sentiment score
def calculate_sentiment(text):
    sentiment = analyzer.polarity_scores(text)
    return sentiment['compound']

# Calculate sentiment scores for each review
df['Sentiment_Score'] = df['cor_lem_review'].apply(calculate_sentiment)

# Filter positive reviews (sentiment score > 0.0)
positive_reviews = df[df['Sentiment_Score'] > 0.60]

# Group the DataFrame by Restaurant and aggregate the positive reviews using ' '.join
positive_reviews['Concatenated_Reviews'] = positive_reviews.groupby('restaurant')['cor_lem_review'].transform(' '.join)

# Drop duplicate rows to get unique restaurant entries
unique_restaurants = positive_reviews.drop_duplicates(subset='restaurant')

# Reset the index and drop the 'Sentiment_Score' column
unique_restaurants = unique_restaurants.reset_index(drop=True).drop(columns=['Sentiment_Score'])

print(unique_restaurants)
```

```python
unique_restaurants.to_csv('concatenated_reviews_data.csv', index=False)

print("DataFrame saved to 'concatenated_reviews_data.csv'")
```
